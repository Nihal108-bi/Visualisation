```python
#Question 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap. 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility

# Define mean and covariance matrix for 3 correlated features
mean = [0, 0, 0]  # Mean of the distributions
cov = [[1, 0.8, 0.5],  # Covariance matrix to induce correlation
       [0.8, 1, 0.6],
       [0.5, 0.6, 1]]

# Generate 100 samples of 3 correlated features
data = np.random.multivariate_normal(mean, cov, 100)

# Create a DataFrame
df = pd.DataFrame(data, columns=['Feature_1', 'Feature_2', 'Feature_3'])
print(df)

# Step 2: Calculate the correlation matrix
corr_matrix = df.corr()

# Step 3: Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=1, linecolor='black')

# Add title
plt.title('Correlation Matrix of Features')

# Show the plot
plt.show()
```

```python
#Question 3.Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
# np.random.seed(20) 
# data = { 
#     'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100), 
#     'Day': np.random.choice(range(1, 31), 100), 
#     'Sales': np.random.randint(1000, 5000, 100) 
# } 
# df = pd.DataFrame(data)


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate the sales data
np.random.seed(20)
data = { 
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100), 
    'Day': np.random.choice(range(1, 31), 100), 
    'Sales': np.random.randint(1000, 5000, 100) 
} 
df = pd.DataFrame(data)

# Step 2: Pivot the data to create a matrix for the heatmap
sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean', fill_value=0)
print(sales_pivot)
# Step 3: Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=0.5)

# Add labels and title
plt.title('Heatmap of Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()

```

```python
#Question 4. Using the given x and y data, generate a 3D surface plot to visualize the function Z=sin(x^2 + y^2)^(1/2)
# x = np.linspace(-5, 5, 100) 
# y = np.linspace(-5, 5, 100) 
# x, y = np.meshgrid(x, y) 
# z = np.sin(np.sqrt(x**2 + y**2))  
# data = {     
#     'X': x.flatten(), 
#     'Y': y.flatten(), 
#     'Z': z.flatten() 
# } 
# df = pd.DataFrame(data)

import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Step 1: Generate the x, y, and z data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Step 2: Flatten the data into a DataFrame
data = {     
    'X': x.flatten(), 
    'Y': y.flatten(), 
    'Z': z.flatten() 
} 
df = pd.DataFrame(data)

# Step 3: Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Step 4: Customize the plot layout
fig.update_layout(title='3D Surface Plot of Z=sin(sqrt(x^2 + y^2))',
                  scene = dict(
                      xaxis_title='X Axis',
                      yaxis_title='Y Axis',
                      zaxis_title='Z Axis'),
                  width=800, height=600)

# Step 5: Show the plot
fig.show()
```

```python
#Question 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x. 
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 1: Prepare the data
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)                # Sine of x

# Step 2: Create a Bokeh plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)

# Step 3: Add a line to the plot
p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

# Step 4: Display the plot in the notebook
output_notebook()
show(p)

 
```

```python
#Question 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns. 
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis256

# Step 1: Generate random data
np.random.seed(42)
x = np.random.rand(100) * 10  # Random values between 0 and 10
y = np.random.rand(100) * 10  # Random values between 0 and 10
sizes = np.random.rand(100) * 20 + 10  # Random sizes between 10 and 30
colors = np.random.rand(100) * 100  # Random values for colors

# Step 2: Create a ColumnDataSource to store the data
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Step 3: Create a scatter plot
p = figure(title="Scatter Plot with Varying Sizes and Colors", width=700, height=400, x_axis_label='X', y_axis_label='Y')

# Step 4: Add the scatter plot with size and color variations
mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

p.scatter('x', 'y', size='sizes', color=mapper, source=source, fill_alpha=0.6, legend_label="Random Data")

# Step 5: Display the plot
output_notebook()
show(p)
```

```python
#Question 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Step 1: Prepare the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Step 3: Create a Bokeh figure for the bar chart
p = figure(x_range=fruits, title="Fruit Counts", 
           toolbar_location=None, tools="", width=600, height=400)

# Step 4: Add bars to the chart
p.vbar(x='fruits', top='counts', width=0.8, source=source, 
       legend_field='fruits', line_color='white', 
       fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Step 5: Customize the chart
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 40
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Step 6: Display the chart
output_notebook()
show(p)
```

```python
#Question 4.Create a Bokeh histogram to visualize the distribution of the given data.  
data_hist = np.random.randn(1000) 
hist, edges = np.histogram(data_hist, bins=30) 

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 1: Generate the data
data_hist = np.random.randn(1000)  # 1000 random values from a normal distribution

# Step 2: Create the histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Step 3: Create a Bokeh figure for the histogram
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', y_axis_label='Frequency', 
           width=700, height=400)

# Step 4: Add quad glyphs to represent the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="white", alpha=0.7)

# Step 5: Display the plot
output_notebook()
show(p)
```

```python
#Question 5. Create a Bokeh heatmap using the provided dataset
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0, 1, 10) 
# y = np.linspace(0, 1, 10) 
# xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Step 1: Generate the heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10) 
y = np.linspace(0, 1, 10) 
xx, yy = np.meshgrid(x, y)

# Step 2: Create a Bokeh figure for the heatmap
p = figure(title="Heatmap Example", x_range=(0, 1), y_range=(0, 1), 
           toolbar_location=None, tools="", width=600, height=600)

# Step 3: Set up the color mapper for the heatmap
color_mapper = LinearColorMapper(palette=Viridis256, low=np.min(data_heatmap), high=np.max(data_heatmap))

# Step 4: Plot the heatmap using the image function
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Step 5: Add a color bar to show the scale
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0), title='Values')
p.add_layout(color_bar, 'right')

# Step 6: Display the heatmap
output_notebook()
show(p)
```
